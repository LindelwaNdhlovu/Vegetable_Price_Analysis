### 5.2 Outliers <a class="anchor" id="section_1_1"></a>

>Due to the very strange price variation observed above for Methi and Green Chilli, we will construct box plots to detect any outliers

```python
# Plotting box plots
plt.figure(figsize=(10, 6))
data.boxplot()
plt.xticks(rotation=45)
plt.title("Box Plot for Vegetable Prices")
plt.show()
```

From the box plot above it is clear that there is an extreme outlier with the Methi Vegetable price. We constructed a histogram to condirm this observation:

```python
# Plotting histograms for each vegetable price.
plt.figure(figsize=(14, 10))
for i, column in enumerate(data.columns[1:], start=1):  # Exclude date column from visualization
    plt.subplot(3, 4, i)
    sns.histplot(data[column], bins=20, kde=True)
    plt.title(column)
    plt.xlabel('Price')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The histogram confirms the results of the box plot as the plot is greatly skewed to the right, hinting a possible outlier. 

To evaluate any large values in each vegetable we conducted a z-score and manually observed the values.

```python
#create data frame with only numerical data
numerical_df = data.select_dtypes(include = 'number')# Calculate Z-scores
z_scores = numerical_df.apply(zscore)

# Identify outliers
outliers = (z_scores > 3) | (z_scores < -3)
print(numerical_df[outliers.any(axis=1)])
```

The z-score test has picked up a few outliers. However, through inference, we concluded that most of these are not outliers, just really high market prices. This is with the exception of 2000, which is the only price in the 1000s range.

As such, we decided to remove the that value (replace it with N/A) and replot the box plots and histograms.

```python
#replace 2000 with N/A
data['Methi'] = data['Methi'].replace(2000, pd.NA)
```

To plot a histogram we will have to remove rows with N/A.

```python
#drop row with missing value
data_clean = data.dropna()
```

### 5.4 Line Graph for Monthly Vegetable Prices Trend <a class="anchor" id="section_1_1"></a>

>Line graphs are used to track changes in vegetable prices over time. This visualization shows the trend of prices for each vegetable month by month, revealing any seasonal patterns or trends.

```python
# Group by month and calculate the mean prices
monthly_prices = data_clean.groupby(data_clean['Price Dates'].dt.to_period('M')).mean()

# Set the Date column as the index
monthly_prices.set_index('Price Dates', inplace=True)

# Plotting the monthly trends for each vegetable
plt.figure(figsize=(14, 8))

# Loop through each vegetable column and plot
for column in monthly_prices.columns:
    plt.plot(monthly_prices.index, monthly_prices[column], label=column)

# Customize the plot
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price Time Series for Multiple Columns')
plt.legend()
plt.grid(True)
plt.show()
```

From the time series analysis it can be conluded that garlic has seen an increase in prices over the year, while the other vegetables, with the exception of peas and Green Chilli, have been steady. 

Pease has an obnormal price change pattern, with sudden increases and decreases throughout the year. 

Green Chilli prices are steady outside of the winter months, which shows a possible increase in demand in winter. 

### 5.5 Bar Plot for Monthly Average Vegetable Prices <a class="anchor" id="section_1_1"></a>

>Bar plots compare the average prices of vegetables across different months. This visualization is useful for identifying which months have higher or lower average prices for each vegetable.

```python
# Bar plot for monthly averages
plt.figure(figsize=(14, 8))

# Define colors for each vegetable
colors = ['#19305B', '#D32F2F', '#8E24AA', '#FBC02D', '#1976D2', '#388E3C', '#F57C00', '#7B1FA2', '#0097A7']

# Extend colors if there are more vegetables than colors provided
if len(colors) < len(monthly_prices.columns):
    additional_colors = plt.cm.tab10(np.linspace(0, 1, len(monthly_prices.columns) - len(colors)))
    colors.extend([plt.cm.colors.rgb2hex(color) for color in additional_colors])

# Loop through each vegetable column and plot with custom colors
for i, column in enumerate(monthly_prices.columns):
    plt.bar(monthly_prices.index.strftime('%b %Y'), monthly_prices[column], label=column, alpha=0.7, color=colors[i])

# Customize plot
plt.title('Monthly Average Vegetable Prices')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```

From the bar plot it can be concluded that garlic has the highest prices for each month, followed by Peas for the months of April to Nov. The months where peas prices are not prominent, the other vegetables are equally distributed.
